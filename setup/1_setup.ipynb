#### Environment erzeugen

Wir arbeiten in Python mit *Environments*. Ein Environment ist ein Kontainer für Python Module. Python Module sind letztlich einfach nur eine Ansammlung von Klassen und Funktionen, die extern bereitgestellt werden und bei der Programmierung verwendet werden können  (wird in anderen Programmiersprachen oftmals auch als *Library* bezeichnet). Z.B. liefert uns das Modul `Crypto` viele Funktionen, die wir im Rahmen unseres Projekts verwenden werden. Weil wir aber nicht für jedes beliebige Projekt z.B. einen Hashfunktion benötigen, macht es aus Ressourcengründen Sinn dieses Modul nur in dem zu unserem Projekt gehörenden Environment zu installieren. Außerdem gibt es für Python Module verschiedene Versionen, wir verwenden z.B. die Version 2.6.1 des `Crypto` Moduls. Da bei der Entwicklung vieler Module die Abwärts-/Aufwärtskompatibilität nicht unbedingt gewährleistet ist, kann es z.B. sein, dass der von uns entwicklete Code mit einer älteren/neueren Version dieses Moduls nicht funktioniert. Gleichzeitig kann es sein, dass wir diese ältere/neuere Version des Moduls jedoch in einem anderen Projekt benötigen, das zu einem anderen Zeitpunkt entwickelt wurde. Damit wir nicht jedes mal wenn wir zwischen den zwei Projekten wechseln die eine Version deinstallieren und die andere installieren müssen, macht es Sinn in getrennten Environments zu arbeiten in denen die jeweils passenden Version des Moduls installiert ist.  
Ein weitere Vorteil von Environments besteht in der Möglichkeit diese zu importieren/exportieren um ein bestimmtes Programm z.B. nach der Entwicklung auf einem anderen Rechner auszuführen ohne zuvor den hier beschriebenen Setup-Prozess komplett durchlaufen zu müssen.

1. Öffne eine Kommandozeile und gebe folgenden Befehl ein:

    `conda env list`
    
   Ihr bekommt eine Liste (in diesem Fall mit nur einem Eintrag) mit den installierten Environments zurück. Hier sollte aktuell nur das *base* (heißt manchmal auch *root*) Environment installiert sein. Der Stern zeigt an, dass dies das aktuell aktive Environment ist. Aktiv bedeutet lediglich, dass dieser Kernel (manchmal verwenden wir den Begriff *Kernel* als Synonym für *Environment*) aktuell verwendet werden würde, wenn wir ein Python Programm (ohne spez. Kernelparameter) über die Kommandozeile ausführen würdet.
2. Erzeuge mit folgendem Befehl ein neues Python 3.6 Environment mit dem Namen fed_coin:

    `conda create --name fed_coin  python=3.6`    
    Bestätigt die Abfrage mit der Eingabe von `y`. Die Warnung kann ignoriert werden.  
3. Aktiviert dieses Environment:<br/>
    *Windows*:
    `activate fed_coin`
    <br/>
    *macOS*:
    `source activate fed_coin`
    <br/>
    Es sollte nun zu Beginn der Kommandozeile `(fed_coin)` stehen.

    (Um das Environment wieder zu deaktivieren lauten die Befehle Windows: `deactivate`, macOS:`source deactivate`)

#### Module installieren

Nun sollte das `fed_coin` Environment aktiv sein. Im nächsten Schritt installieren wir alle Module, die wir für unser Projekt benötigen, in dieses Environment.

1. Um zusehen welche Module im aktuell aktiven Environment installiert sind, kann folgender Befehl verwendet werden:

    `conda list`<br/>
    <br/>
2. Für die Installation verwenden wir den Modulmanager `pip`. Als erstes muss `pip` auf den neusten Stand gebracht werden. Verwende dazu folgenden Befehl:

    `python -m pip install --upgrade pip`<br/>
    <br/>
3. Um z.B. das Modul `Crypto` zu installieren, gebe folgenden Befehl in die Kommandozeile ein:

     `pip install pycryptodome`<br/>
    <br/>
    *Zusätzlicher Hinweise:*  
    a. Manchmal weicht der Modulname in `Python` vom Installationsname in `pip` ab, hier z.B. `Crypto` vs. `pycryptodome`.  
    b. Falls im Rahmen der Installation eines Moduls Fragen in der Kommandozeile auftauchen, sind die i.d.R. mit `y` zu bestätigen.

    *Außerdem:*
    Möchte man eine spezezielle Version des Moduls installieren, so geht das mit einem Befehl der z.B. so aussieht:

     `pip install pycryptodome==3.6.6`<br/> 
     <br/>
    Wir verwenden allerdings immer die aktuellste Version der Module, deshalb sind die Versionsangaben für unsere Anwendung überflüssig.
    <br/>
4. Installiere analog zum `Crypto` Modul mit Hilfe von `pip` folgende Module in das `fed_coin` Environment:

    * `requests`
    * `flask`
    * `ipykernel`

    <br/>
5. Deaktiviere das `fed_coin` Environment per <br/>
*Windows*: `deactivate`<br/> 
*macOS*: `source deactivate`<br/>
--> Jetzt ist automatisch das `base` Env aktiv.
<br/>
6. Führe (im base Env, welches jetzt automatisch aktiv ist) folgenden Befehl aus:

    `conda install nb_conda_kernels`
    <br/>
    Dadurch wird per `conda` Modulmanager (ähnlich wie `pip`) das `nb_conda_kernels` Modul im `base` Env installiert. Wir brauchen dieses Modul um den `fed_coin` Kernel später aus unseren `jupyter` notebooks aufzurufen.  
    *Hinweis für Windows: Ggf. erscheint eine Warnung, dass das Modul nur nach einer Aktualisierung von Conda installiert werden kann. Wenn die Aktualisierung nicht erfolgreich ist, dann muss die Konsole geschlossen und erneut mit Admin-Rechten geöffnet werden (Per Rechtsklick auf das Konsolensymbol und  dann "Als Administrator öffnen" auswählen). Nur mit der Berechtigung kann das Update in diesem Fall ausgeführt und im Anschluss das Modul installiert werden.*
